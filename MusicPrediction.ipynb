## **Import Data**

```python
import pandas as pd
music_data = pd.read_csv('music.csv')
music_data
```

## **Prepare Data**

### **Split Data Into X and Y**

```python
X = music_data.drop(columns=['genere'])
print(X)
Y = music_data['genere']
print(Y)
```

## **Split Data Into Training and Testing**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
print(X_train)
```

## **Build Model**

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

## **Make Prediction**

```python
prediction = model.predict(X_test)   # here  Y_test is for checking accuracy. for checking how accuratly model predict y on test data
print("Prediction","Actual")
for i in range(len(prediction)):
  print( prediction[i],  Y_test.iloc[i])

pre = model.predict([[36,1]])
print(pre)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(Y_test, prediction)
score
```
